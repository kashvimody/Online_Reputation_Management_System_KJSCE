```python
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

from selenium.webdriver.chrome.service import Service
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service)

# Open Quora
driver.get('https://www.quora.com/')
time.sleep(3)  # Wait for the page to load

# Ensure the page has fully loaded
WebDriverWait(driver, 180).until(
    lambda d: d.execute_script('return document.readyState') == 'complete'
)

# Wait until the search box becomes clickable
search_box = WebDriverWait(driver, 180).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, 'input[placeholder="Search Quora"]'))
)

search_box.click()
search_box.send_keys('K.J. Somaiya College of Engineering')
search_box.send_keys(Keys.RETURN)
time.sleep(5)

# Scroll down to load more results
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

# Extract question elements using the specific class from the HTML you provided
questions = driver.find_elements(By.XPATH, "//div[contains(@class, 'puppeteer_test_question_title')]")

# Prepare to store results in CSV
csv_file = 'quora_questions.csv'
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Question', 'Link'])

    # Extract the question text and the associated link
    for question in questions:
        try:
            # Get the question text
            title = question.text.strip()  # Get the question text and remove leading/trailing whitespace

            # Check if the question title is not empty
            if title:
                print(f"Question: {title}")
                
                # Find the ancestor 'a' tag to get the question link
                try:
                    link_element = question.find_element(By.XPATH, ".//ancestor::a[1]")
                    link = link_element.get_attribute('href')
                    print(f"Link: {link}")
                    
                    # Save the question and link to the CSV file
                    writer.writerow([title, link])

                except Exception as e:
                    print(f"Error finding link for question: {title}")
                    print(f"Error: {e}")
            print('---')
        except Exception as e:
            print(f"Error processing question: {e}")

# Close the browser
driver.quit()
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set up ChromeDriver using Service
from selenium.webdriver.chrome.service import Service
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service)

# Open the CSV file with questions and links
with open('quora_questions.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Prepare to store the answers in a new CSV file
    with open('quora_answers.csv', mode='w', newline='', encoding='utf-8') as answerfile:
        writer = csv.writer(answerfile)
        writer.writerow(['Question', 'Link', 'Answer'])

        # Iterate over each row in the CSV file
        for row in reader:
            question = row['Question']
            link = row['Link']
            
            # Visit the question page
            driver.get(link)
            time.sleep(5)  

            # Extract the top answer 
            try:
                answer_element = driver.find_element(By.XPATH, "//div[@class='q-box spacing_log_answer_content puppeteer_test_answer_content']")
                answer = answer_element.text.strip()
                print(f"Answer for '{question}': {answer}")

                # Write the question, link, and answer to the new CSV
                writer.writerow([question, link, answer])

            except Exception as e:
                print(f"No answer found for question: {question}")
                print(f"Error: {e}")
                writer.writerow([question, link, "No answer found"])

# Close the browser
driver.quit()
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize the ChromeDriver
from selenium.webdriver.chrome.service import Service
service = Service('/usr/local/bin/chromedriver')
driver = webdriver.Chrome(service=service)

# Open the existing CSV with questions and links
with open('quora_questions.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Prepare to store the answers in a list
    updated_data = []

    for row in reader:
        question = row['Question']
        link = row['Link']
        answer = 'No answer found'  # Default answer in case we can't find it
        
        # Visit the question page
        driver.get(link)
        time.sleep(5)  # Give the page time to load

        # Check for the "Continue Reading" button and click it if present
        try:
            continue_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//div[text()='Continue Reading']"))
            )
            continue_button.click()
            time.sleep(3)  # Wait for the full answer to load
        except Exception as e:
            print(f"No 'Continue Reading' button for question: {question}")

        # Try to directly extract the full answer after clicking "Continue Reading"
        try:
            answer_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//div[@class='q-box spacing_log_answer_content puppeteer_test_answer_content']"))
            )
            full_answer = answer_element.get_attribute("innerText").strip()
            print(f"Full Answer for '{question}': {full_answer}")
            answer = full_answer
        except Exception as e:
            print(f"Could not find a full answer for question: {question}")
            print(f"Error: {e}")

        # Append the question, link, and full answer to the list
        updated_data.append({'Question': question, 'Link': link, 'Answer': answer})

# Close the browser
driver.quit()

# Write the updated data (with full answers) back to the CSV
with open('quora_questions_with_full_answers_automate.csv', mode='w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Question', 'Link', 'Answer']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()  # Write header
    writer.writerows(updated_data)  # Write the updated data with full answers

print("Updated CSV with full answers has been saved as 'quora_questions_with_full_answers_automate.csv'.")
```

```python
import pandas as pd
from textblob import TextBlob
from IPython.display import display

# Load the CSV file
file_path = 'final_quora_cleaned_data.csv'
df = pd.read_csv(file_path)

# Check the contents of the file
print("First few rows of the original data:")
display(df.head())

# Define a function to calculate sentiment using TextBlob
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment
    return sentiment.polarity, sentiment.subjectivity

# Apply the sentiment function to the DataFrame
df['polarity'], df['subjectivity'] = zip(*df['Answer'].apply(get_sentiment))

# Convert polarity to categories
def categorize_polarity(polarity):
    if polarity > 0.1:
        return 'positive'
    elif polarity < -0.1:
        return 'negative'
    else:
        return 'neutral'

# Convert subjectivity to categories (optional)
def categorize_subjectivity(subjectivity):
    return 'subjective' if subjectivity > 0.5 else 'objective'

# Apply the categorization functions
df['polarity_category'] = df['polarity'].apply(categorize_polarity)
df['subjectivity_category'] = df['subjectivity'].apply(categorize_subjectivity)

# Save the updated DataFrame with sentiment scores and categories to a new CSV file
output_path = 'sentiment_analysis_output_trial.csv'
df.to_csv(output_path, index=False)

print(f'Sentiment analysis results saved to {output_path}')

# Display the results
print("First few rows of the updated data with sentiment analysis:")
display(df[['Answer', 'polarity', 'subjectivity', 'polarity_category', 'subjectivity_category']].head())
```
